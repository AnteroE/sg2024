```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from collections import Counter

# Load the data
# Assuming the data is in a CSV file named 'blockchain_data.csv'
data = pd.read_csv('decoded_with_headers.csv')

# Step 1: Feature Engineering
def engineer_features(df):
    # Group by from_address and calculate various metrics
    address_features = df.groupby('from_address').agg({
        'block_number': ['count', 'min', 'max', 'std'],
        'transaction_index': ['mean', 'std'],
        'log_index': ['mean', 'std'],
        'event_name': lambda x: Counter(x).most_common(1)[0][0]  # Most common event
    })
    
    # Flatten column names
    address_features.columns = ['_'.join(col).strip() for col in address_features.columns.values]
    
    # Calculate time span of activity
    address_features['block_span'] = address_features['block_number_max'] - address_features['block_number_min']
    
    # Analyze non-None parameters
    for i in range(1, 11):
        param_col = f'param_{i}'
        if param_col in df.columns:
            address_features[f'{param_col}_unique_count'] = df.groupby('from_address')[param_col].nunique()
    
    return address_features

# Step 2: Clustering
def cluster_addresses(features):
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    
    dbscan = DBSCAN(eps=0.5, min_samples=5)
    clusters = dbscan.fit_predict(scaled_features)
    
    return clusters

# Step 3: Analyze Clusters
def analyze_clusters(df, features, clusters):
    features['cluster'] = clusters
    cluster_summary = features.groupby('cluster').agg({
        'block_number_count': ['mean', 'std'],
        'block_span': ['mean', 'std'],
        'event_name': lambda x: Counter(x).most_common(3)
    })
    
    # Identify potential Sybil clusters
    potential_sybil_clusters = cluster_summary[
        (cluster_summary[('block_number_count', 'std')] < cluster_summary[('block_number_count', 'std')].median()) &
        (cluster_summary[('block_span', 'std')] < cluster_summary[('block_span', 'std')].median())
    ]
    
    return cluster_summary, potential_sybil_clusters

# Main execution
def main():
    # Engineer features
    address_features = engineer_features(data)
    
    # Perform clustering
    clusters = cluster_addresses(address_features.drop('event_name', axis=1))
    
    # Analyze clusters
    cluster_summary, potential_sybil_clusters = analyze_clusters(data, address_features, clusters)
    
    print("Cluster Summary:")
    print(cluster_summary)
    print("\nPotential Sybil Clusters:")
    print(potential_sybil_clusters)
    
    # Further analysis on potential Sybil clusters
    for cluster in potential_sybil_clusters.index:
        cluster_addresses = address_features[address_features['cluster'] == cluster].index
        cluster_data = data[data['from_address'].isin(cluster_addresses)]
        
        print(f"\nDetailed Analysis for Cluster {cluster}:")
        print(cluster_data['event_name'].value_counts(normalize=True))
        print(cluster_data['block_number'].describe())

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

def load_data(filename):
    df = pd.read_csv(filename, on_bad_lines='skip')
    print(f"Loaded {df.shape[0]} rows and {df.shape[1]} columns")
    print("\nColumn names:")
    print(df.columns.tolist())
    print("\nMissing values:")
    print(df.isnull().sum())
    return df

def clean_data(df):
    # Drop rows with missing values
    df = df.dropna()
    print(f"\nCleaned data shape: {df.shape}")
    return df

def entropy(probabilities):
    # Remove zero probabilities to avoid log2(0)
    probabilities = probabilities[probabilities > 0]
    return -(probabilities * np.log2(probabilities)).sum()

def engineer_features(df):
    # Group by from_address and calculate various metrics
    address_features = df.groupby('from_address').agg({
        'block_number': ['count', 'min', 'max', 'std'],
        'transaction_index': ['mean', 'std'],
        'log_index': ['mean', 'std'],
    })
    
    # Flatten column names
    address_features.columns = ['_'.join(col).strip() for col in address_features.columns.values]
    
    # Calculate time span of activity
    address_features['block_span'] = address_features['block_number_max'] - address_features['block_number_min']
    
    # Replace zero block_span with NaN to avoid division by zero
    address_features['block_span'].replace(0, np.nan, inplace=True)
    
    # Calculate transaction frequency
    address_features['transaction_frequency'] = address_features['block_number_count'] / address_features['block_span']
    
    # Replace inf values resulting from division by zero
    address_features['transaction_frequency'].replace([np.inf, -np.inf], np.nan, inplace=True)
    
    # Add most common event for each address
    most_common_events = df.groupby('from_address')['event_name'].agg(lambda x: Counter(x).most_common(1)[0][0])
    address_features['most_common_event'] = most_common_events
    
    # Analyze non-None parameters
    for i in range(1, 11):
        param_col = f'param_{i}'
        if param_col in df.columns:
            # Compute counts for each parameter value per address
            counts = df.groupby('from_address')[param_col].value_counts()
            unique_counts = counts.groupby('from_address').size()
            # Prepare counts for entropy calculation
            probs = counts.groupby('from_address', group_keys=False).apply(lambda x: x / x.sum())
            entropies = probs.groupby('from_address').apply(entropy)
            address_features[f'{param_col}_unique_count'] = unique_counts
            address_features[f'{param_col}_entropy'] = entropies
    
    return address_features

def cluster_addresses(features):
    # Standardize features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    
    # Perform DBSCAN clustering
    clusterer = DBSCAN(eps=0.5, min_samples=5)
    clusterer.fit(scaled_features)
    
    # Extract cluster labels
    clusters = clusterer.labels_
    
    return clusters

def visualize_clusters(features, clusters):
    # For visualization, reduce dimensions to 2D using PCA
    from sklearn.decomposition import PCA

    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(features)

    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(principal_components[:, 0], principal_components[:, 1], c=clusters, cmap='viridis', s=50)
    plt.legend(*scatter.legend_elements(), title="Clusters")
    plt.title('DBSCAN Clustering Visualization')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()

def analyze_clusters(data, address_features, clusters):
    # Add cluster labels to address_features
    address_features['cluster'] = clusters

    # Analyze clusters
    cluster_counts = address_features['cluster'].value_counts()
    cluster_summary = cluster_counts.describe()
    
    # Identify potential Sybil clusters (clusters with more than one address)
    potential_sybil_clusters = cluster_counts[cluster_counts > 1]
    
    return cluster_summary, potential_sybil_clusters

def main():
    raw_data = load_data('decoded_with_headers (1).csv')
    data = clean_data(raw_data)
    address_features = engineer_features(data)
    
    # Remove non-numeric columns before clustering
    clustering_features = address_features.select_dtypes(include=[np.number])

    # Replace infinite values and NaNs
    clustering_features.replace([np.inf, -np.inf], np.nan, inplace=True)
    clustering_features.fillna(0, inplace=True)
    
    clusters = cluster_addresses(clustering_features)
    
    visualize_clusters(clustering_features, clusters)
    
    cluster_summary, potential_sybil_clusters = analyze_clusters(data, address_features, clusters)
    
    print("\nCluster Summary:")
    print(cluster_summary)
    print("\nPotential Sybil Clusters:")
    print(potential_sybil_clusters)
    
    for cluster in potential_sybil_clusters.index:
        addresses_in_cluster = address_features[address_features['cluster'] == cluster].index
        cluster_data = data[data['from_address'].isin(addresses_in_cluster)]
        print(f"\nDetailed Analysis for Cluster {cluster}:")
        print(cluster_data['event_name'].value_counts(normalize=True))
        print(cluster_data['block_number'].describe())
        
        # Additional analysis
        print("\nUnique addresses in cluster:", len(addresses_in_cluster))
        print("Most common events:")
        print(cluster_data['event_name'].value_counts().head())
        
        # Visualize event distribution
        plt.figure(figsize=(10, 6))
        cluster_data['event_name'].value_counts().plot(kind='bar')
        plt.title(f'Event Distribution for Cluster {cluster}')
        plt.xlabel('Event Name')
        plt.ylabel('Count')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

if __name__ == "__main__":
    main()
```
